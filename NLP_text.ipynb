```python
filtered_list=[]
def text_classification(text):
    remove_specialChar=re.sub("[^A-Za-z]+"," ",text) #remove the sp_char
    sent_token=sent_tokenize(remove_specialChar)# tokenize sentence
    word_token=word_tokenize(remove_specialChar)#tokenize words
    stop_words=set(stopwords.words('english'))#show stop words
    for word in word_token:             
          if word not in stop_words:          #exclusive stop_words
                filtered_list.append(word)
                tagged=nltk.pos_tag(filtered_list)#tagging pos
                
    
    return tagged   
```

```python
result=text_classification(pageObj.extractText())
result
```
